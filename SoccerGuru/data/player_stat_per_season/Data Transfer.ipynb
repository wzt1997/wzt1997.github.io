```python
import csv
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv("player_stat.csv", encoding='utf-8')
```

```python
players = df.to_dict(orient='records')
df.columns.values.tolist()
```

```python
player_data = { }

def getPartPlayerStat( p ):
    stat_names = { 'goals','assists','goalsAssistsSum',\
                  'yellowCards','directRedCards','redCards','minutesPlayed','matchesStarted',\
                  'id','type','appearances','rating','totalPasses','keyPasses','tackles',\
                  'interceptions','totalShots','shotsOnTarget','shotsOffTarget','clearances',\
                  'dispossessed','wasFouled','fouls','team>name','team>nameCode','team>ranking',\
                  'team>id','player_name','tournament_name','year' }
    new_p = {key: value for key, value in p.items() if key in stat_names}
    return new_p

for p in players:
    name = p['player_name']
    tourna = p['tournament_name']
    if not name in player_data:
        player_data[name] = {}
    if not tourna in player_data[name]:
        player_data[name][tourna] = []
    player_data[name][tourna].append( getPartPlayerStat(p))

# filter out the players that have too few records
player_data_copy = player_data.copy()
for name in player_data_copy.keys():
    cnt = 0
    for tourna in player_data[name].keys():
        cnt += len(player_data[name][tourna])
    if cnt < 4:
        del player_data[name]
```

```python
with open("player_data.json", "w") as f:
    f.write(json.dumps(player_data, ensure_ascii=False, indent=4, separators=(',', ':')))
```

```python
pie_data = {}
pie_data_field = [ 2, 2, 3, 2, 2]

for p in players:
    name = p['player_name']
    year = p['year']
    if not name in pie_data:
        pie_data[name] = {}
    if not year in pie_data[name]:
        pie_data[name][year] = [[{'x':'shotsOnTarget', 'y': 0}, {'x':'shotsOffTarget', 'y': 0 }], \
                                [{'x':'goalsFromInsideTheBox', 'y': 0}, {'x':'goalsFromOutsideTheBox', 'y': 0}], \
                                [{'x':'headedGoals', 'y': 0}, {'x':'leftFootGoals', 'y': 0}, {'x':'rightFootGoals', 'y': 0}], \
                                [{'x':'accuratePasses', 'y': 0}, {'x':'inaccuratePasses', 'y': 0}], \
                                [{'x':'aerialDuelsWon', 'y': 0}, {'x':'groundDuelsWon', 'y': 0}]]
    for index in range(0,5):
        element = pie_data_field[index]
        for i in range(0, element):
            tmp = pie_data[name][year][index][i]
            tmp['y'] += p[tmp['x']]    

# filter out the players that have too few records
pie_data_copy = pie_data.copy()
for name in pie_data_copy.keys():
    if not name in player_data.keys():
        del pie_data[name]
```

```python
with open("pie_data.json", "w") as f:
    f.write(json.dumps(pie_data, ensure_ascii=False, indent=4, separators=(',', ':')))
```
